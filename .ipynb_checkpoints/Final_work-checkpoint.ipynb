```python

import numpy as np
import numpy.random as nr
from numpy.linalg import norm
import plotly.express as px
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

## Problema

**Contexto comercial.** La reincidencia es una medida de la probabilidad de que un ex recluso sea arrestado nuevamente, condenado nuevamente o regresado a prisión con o sin una nueva sentencia durante un período de tres años después de la liberación del recluso. Se ha utilizado para estudiar el desempeño y la eficacia de las prisiones de gestión pública y privada. Las tasas de reincidencia entre los presos en los Estados Unidos son altas, con más del 60% de los presos liberados que reinciden dentro de los tres años posteriores a la liberación.

En consecuencia, los programas de reforma penitenciaria han invertido cientos de millones de dólares por año en programas diseñados para reducir la reincidencia. Sin embargo, implementar estos programas es costoso y evaluarlos puede ser complejo. Actualmente, los legisladores están interesados en evaluar un programa de capacitación diseñado para reducir la reincidencia entre los reclusos en libertad condicional.

```python
df = pd.read_csv('rec_study.csv')
df
```

| Variable | Descripción|
|----|------------------|
|ID  | Identificador Prisionero|
|Max severity | (1-5) Gravedad del cargo más grave en la condena actual |
|Ave severity | (1-5) Severidad promedio en todos los cargos en la condena actual | 
|Sentence time| Duración de la sentencia (meses) asignada |
|Sentence served| Tiempo cumplido (meses).| 
| Education | (1-4) Organizado de forma que: 1 - Menor a secundaria, 2 - Secuendaria, 3 - Algo de universidad y 4 - Título universitario |  
| Repeat offender| Indicador binario de Reincidencia | 
| Substance use | Indicador binario de abuso de sustancias | 
| Support | Indicador binario de que la persona en libertad condicional está recibiendo apoyo de transición (p. ej., centro de transición) |

```python
#Definición de valores x y y
X = df[['max_severity', 'sentence_time', 'sentence_served', 'substance_use', 'support', 'education' ]]
y = df[['reoffend']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_predict = dtc.predict(X_test)
print('The accuracy of Decision Tree Classification is', dtc.score(X_test, y_test))
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_y_pred = rfc.predict(X_test)
print('The accuracy of random forest classifier is', rfc.score(X_test, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
gbc_y_pred = gbc.predict(X_test)
print('The accuracy of gradient tree boosting is', gbc.score(X_test, y_test))
```

## Optimización DTC

```python
param_grid = {'max_features': ['auto', 'sqrt', 'log2'],
              'ccp_alpha': [0.1, .01, .001],
              'max_depth' : [5, 6, 7, 8, 9],
              'criterion' :['gini', 'entropy']
             }
tree_clas = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(estimator=tree_clas, param_grid=param_grid, cv=5, verbose=True)
grid_search.fit(X_train, y_train)
```

```python
final_model = grid_search.best_estimator_
final_model
```

```python
tree_clas = DecisionTreeClassifier(ccp_alpha=0.001, max_depth=9, max_features='auto',
                       random_state=42)
tree_clas.fit(X_train, y_train)
y_predict = tree_clas.predict(X_test)
```

```python
print('The accuracy of Decision Tree Classification is', tree_clas.score(X_test, y_test))
```

## Optimización RFC

```python
rfc=RandomForestClassifier(random_state=42)
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train)
```

```python
final_model_rfc = CV_rfc.best_estimator_
final_model_rfc
```

```python
rf_clas = RandomForestClassifier(criterion='entropy', max_depth=8, n_estimators=500,
                       random_state=42)
rf_clas.fit(X_train, y_train)
y_predict = rf_clas.predict(X_test)
```

```python
print('The accuracy of random forest classifier is', rf_clas.score(X_test, y_test))
```

<h1 style="font-size:20px">Optimización XGB</h1>

```python
parameters = {
    "loss":["deviance"],
    "learning_rate": [0.01, 0.05, 0.1, 0.2],
    "min_samples_split": np.linspace(0.1, 0.5, 2),
    "min_samples_leaf": np.linspace(0.1, 0.5, 2),
    "max_depth":[3,5],
    "max_features":["log2","sqrt"],
    "criterion": ["friedman_mse",  "mae"],
    "subsample":[0.5, 0.8, 0.9, 1.0],
    "n_estimators":[10]
    }
clf = GridSearchCV(GradientBoostingClassifier(), parameters, cv=8, n_jobs=-1)

clf.fit(X_train, y_train)
```

```python
final_model_gbc = clf.best_estimator_
final_model_gbc
```

```python
gb = GradientBoostingClassifier(learning_rate=0.2, max_depth=5, max_features='log2',
                           min_samples_leaf=0.1, min_samples_split=0.1,
                           n_estimators=10,
                       random_state=42)
gb.fit(X_train, y_train)
y_predict = gb.predict(X_test)
```

```python
print('The accuracy of gradient tree boosting is', gb.score(X_test, y_test))
```

<h1 style="font-size:20px">Comparación de modelos</h1>

```python
from sklearn.metrics import classification_report, accuracy_score

# Predicciones del modelo 1
y_pred_tree_clas = tree_clas.predict(X_test)
print("tree_clas")
print(classification_report(y_test, y_pred_tree_clas))
print("Accuracy:", accuracy_score(y_test, y_pred_tree_clas))

# Predicciones del modelo 2
y_pred_rf_clas = rf_clas.predict(X_test)
print("rf_clas")
print(classification_report(y_test, y_pred_rf_clas))
print("Accuracy:", accuracy_score(y_test, y_pred_rf_clas))

# Predicciones del modelo 3
y_pred_gb = gb.predict(X_test)
print("gb")
print(classification_report(y_test, y_pred_gb))
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
```

<h1 style="font-size:20px">Visualización de modelos</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

# Metricas del modelo 1
tree_clas_report = classification_report(y_test, y_pred_tree_clas, output_dict=True)
tree_clas_precision = tree_clas_report['macro avg']['precision']
tree_clas_recall = tree_clas_report['macro avg']['recall']
tree_clas_f1 = tree_clas_report['macro avg']['f1-score']

# Metricas del modelo 2
rf_clas_report = classification_report(y_test, y_pred_rf_clas, output_dict=True)
rf_clas_precision = rf_clas_report['macro avg']['precision']
rf_clas_recall = rf_clas_report['macro avg']['recall']
rf_clas_f1 = rf_clas_report['macro avg']['f1-score']

# Metricas del modelo 3
gb_report = classification_report(y_test, y_pred_gb, output_dict=True)
gb_precision = gb_report['macro avg']['precision']
gb_recall = gb_report['macro avg']['recall']
gb_f1 = gb_report['macro avg']['f1-score']

# Definimos los datos
precisions = [tree_clas_precision, rf_clas_precision, gb_precision]
recalls = [tree_clas_recall, rf_clas_recall, gb_recall]
f1_scores = [tree_clas_f1, rf_clas_f1, gb_f1]

# Creamos la figura
fig, ax = plt.subplots(figsize=(6,4))

# Definimos la posición de las barras en el eje x
x = np.arange(len(precisions))

# Creamos las barras
ax.bar(x - 0.2, precisions, 0.2, label='Precisión')
ax.bar(x, recalls, 0.2, label='Exhaustividad')
ax.bar(x + 0.2, f1_scores, 0.2, label='F1-Score')

# Agregamos leyendas y título
ax.set_xticks(x)
ax.set_xticklabels(['Modelo 1', 'Modelo 2', 'Modelo 3'])
ax.legend()
ax.set_title('Comparación de métricas de los modelos')

# Mostramos el gráfico
plt.show()

```

<h1 style="font-size:25px">Pipeline</h1>

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Crear el pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Normalización de los datos
    ('classifier', LogisticRegression())  # Clasificador de regresión logística
])

# Entrenamiento del modelo
pipeline.fit(X_train, y_train)

# Evaluación del modelo
accuracy = pipeline.score(X_test, y_test)
print("Accuracy:", accuracy)
```

<h1 style="font-size:25px">Matrix de confusión </h1>

```python
from sklearn.metrics import confusion_matrix

# Crear el pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Normalización de los datos
    ('classifier', LogisticRegression())  # Clasificador de regresión logística
])

# Entrenamiento del modelo
pipeline.fit(X_train, y_train)

# Predicciones
y_pred = pipeline.predict(X_test)

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(2, 2))

# Crear el heatmap
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False)

# Agregar etiquetas
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')

# Mostrar la figura
plt.show()


```

```python
# Cargar el modelo
rfc = joblib.load('rf_clas.joblib')
```

```python
# Definir la función de predicción
def predict(rfc, input_df):
    predictions = rfc.predict(input_df)
    return predictions
```

```python
# Configurar la página de Streamlit
def app():
    st.title('Mi_aplicación_de_clasificacion')
    streamlit Mi_aplicación_de_clasificacion.py

    # Agregar una sección de carga de datos
    st.subheader('Ingrese los datos a clasificar:')
    max_severity = st.slider('Gravedad máxima', 1, 10)
    ave_severity = st.slider('Gravedad media', 1, 10)
    sentence_time = st.slider('Tiempo de condena', 1, 50)
    sentence_served = st.slider('Tiempo de condena cumplido', 1, 50)
    repeat_offender = st.selectbox('Reincidencia', [0, 1])
    substance_use = st.selectbox('Consumo de sustancias', [0, 1])
    support = st.selectbox('Soporte familiar', [0, 1])
    education = st.selectbox('Educación', [0, 1])

    # Crear un DataFrame con los datos ingresados por el usuario
    input_dict = {'max_severity': max_severity, 'ave_severity': ave_severity, 'sentence_time': sentence_time, 'sentence_served': sentence_served, 'repeat_offender': repeat_offender, 'substance_use': substance_use, 'support': support, 'education': education}
    input_df = pd.DataFrame([input_dict])

    # Realizar la predicción
    if st.button('Predecir'):
        output = predict(rfc=rfc, input_df=input_df)
        if output == 0:
            st.write('El individuo no reincidirá.')
        else:
            st.write('El individuo reincidirá.')

if __name__ == '__main__':
    app()
```
